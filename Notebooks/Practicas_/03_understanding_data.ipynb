```python
#### GRAFICO HEATMAP (mapa de calor) - para hallar correlacion de variables (pearson, kendall, spearman) - importante manejar los atipicos o outliers
```

```python
correlation = df.corr(method='pearson').reset_index().melt('index') #este metodo excluye nulos
correlation.columns = ['Var1', 'Var2', 'Corr_index']

correlation.head()
```

```python
chart = alt.Chart(correlation).mark_rect().encode(
    x=alt.X('Var1',title=None),
    y=alt.Y('Var2',title=None),
    color =alt.Color('Corr_index'),
).properties(width=alt.Step(40), height=alt.Step(40))

chart += chart.mark_text(size=10).encode(
    text=alt.Text('Corr_index', format=".2f"),
    color=alt.condition("datum.correlation > 0.5",alt.value('white'),alt.value('black'))
)
```

```python
correlation = df.corr(method='spearman').reset_index().melt('index') #este metodo excluye nulos
correlation.columns = ['Var1', 'Var2', 'Corr_index']

correlation.head()
chart = alt.Chart(correlation).mark_rect().encode(
    x=alt.X('Var1',title=None),
    y=alt.Y('Var2',title=None),
    color =alt.Color('Corr_index'),
).properties(width=alt.Step(40), height=alt.Step(40))

chart += chart.mark_text(size=10).encode(
    text=alt.Text('Corr_index', format=".2f"),
    color=alt.condition("datum.correlation > 0.5",alt.value('white'),alt.value('black'))
)
chart
```

```python
fec_prueba = pd.to_datetime('09-10-2016')
```

```python
df['ESTU_FECHANACIMIENTO'] =pd.to_datetime(df['ESTU_FECHANACIMIENTO'])
```

```python
df['EDAD'] = round((fec_prueba - df['ESTU_FECHANACIMIENTO']).dt.days/365.2425)
```

```python
Value_count = df.isna().sum()
Value_Count_Mayor = Value_count[Value_count > 0]
print(f'Resumen:\r\n',Value_Count_Mayor)
```

```python
avg_edad = df["EDAD"].astype("float").mean()
# Imprimir los resultados 
print("Promedio de edad:", avg_edad)
```
